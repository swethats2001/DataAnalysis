# Analysis Tasks:
1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees. (2 marks)

```python
#Identify distinct team 
df2["Team"].unique()
```

```python
emp = df2['Team'].value_counts()
percentage_split = (emp / len(df2)) * 100
result = pd.DataFrame({'Team': emp.index, 
                       'Count': emp.values, 
                       'Percentage': percentage_split.values})
print(result)
```

2. Segregate employees based on their positions within the company. 

```python
result = df2['Position'].value_counts()
print(result)
plt.figure(figsize=(8, 5))
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99", "#c2c2f0", "#ffb3e6"]
plt.pie(result.values, labels=result.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Distribution of Employees by Position")
plt.show()
```

* Method : The total number of employees are found with respect to the position using value_counts() method grouped by position.
* A pie chart is plotted to show the insights.
* Findings/Insigts : From the above pie chart it is visible that the highest number of employees are present in 'SG' position(22.3%, 102 employees) and lowest in 'C' position(17.2%, 79 employees).

3. Identify the predominant age group among employees. (2 marks)

```python
age_groups = pd.cut(df2['Age'], 
                    bins=[20, 25, 30, 35, 40, 45, 50], 
                    labels=["20-25", "25-30", "30-35", "35-40", "40-45", "45-50"])
age_group_distribution = age_groups.value_counts()
predominant_age_group = age_group_distribution.idxmax()
print(age_group_distribution)
plt.figure(figsize=(10, 6))
age_group_distribution.plot(kind='bar', color='green')
plt.xlabel("Age Groups")
plt.ylabel("Number of Employees")
plt.title("Age Group Distribution")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

* Method:Age group is found using 'cut()' method to segment and sort data values into bins.
* age_group_distribution and predominant_age_group is found using 'value_counts()' of age_groups and 'idxmax()' of age_group_distribution respectievly.
* A bar chart is plotted to find Age Group Distribution.
* Findings/Insigts: Age group '20-25' has highest number of employees(179).

4. Discover which team and position have the highest salary expenditure. (2 marks)

```python
salexp = df.groupby('Team')['Salary'].sum()
highest_salary = salexp.idxmax()
salexpos = df.groupby('Position')['Salary'].sum()
highest_salary_position = salexpos.idxmax()

plt.figure(figsize=(15, 5))
plt.plot(salexp.index, salexp.values, marker='o', linestyle='-', color='b', label="Team Salary Expenditure")
plt.xlabel("Team")
plt.ylabel("Total Salary Expenditure")
plt.title("Salary Expenditure by Team")
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(salexpos.index, salexpos.values, marker='o', linestyle='-', color='g', label="Position Salary Expenditure")
plt.xlabel("Position")
plt.ylabel("Total Salary Expenditure")
plt.title("Salary Expenditure by Position")
plt.legend()
plt.show()
```

* Method:Highest salary with respect to position and team are found using 'sum()' and 'idmax()' functions.
* A line chart is plotted to visualize the findings.
* Findings/Insigts: Team with highest salary expenditure: Cleveland Cavaliers and Position with highest salary expenditure: C
